```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '' :
        result.append(float(row[-1]))
    
plt.hist(result, bins=100, color='r')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
aug = []
for row in data:
    month = row[0].split(".")[1]
    if row[-1] != '':
        if month == '8':
            aug.append(float(row[-1]))
plt.hist(aug, bins=100, color='r')
plt.show()
```
