# Seaborn

## sns.heatmap()
<br>

**data**
   - 시각화에 사용될 2D 데이터.

    
**vmin, vmax**
   - heatmap 색깔의 범위.

    
**annot**
   - 각각 heatmap의 value. True면 value를 나타냄.
    
**linewidths**
   - 셀을 나누는 line의 두께.
    
**fmt**
   - 문자열 포매팅 방식.

```python
sns.heatmap(data.corr(), annot=True, cmap='RdYlGn', linewidths=0.2)
plt.show()
```

```python
sns.heatmap(data.corr(), annot=True, cmap='RdYlGn', linewidths=0.2, fmt='0.2f')
plt.show()
```
