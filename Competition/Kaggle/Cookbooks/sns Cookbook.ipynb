# seaborn.barplot()
막대그래프 시각화.

**Parameter**: x, y, data, palette, orient, \**xerr

- orient: 그래프 타입이 v(수직)인지 h(수평)인지.
- xerr: 표준편차 막대기(errorbar) 표시.

```python
# 예제
g = sns.barplot('CrossValMeans', 'Algorithm', data=cv_res, 
               palette='Set3', orient='h', **{'xerr': cv_std})
```

# sns.pairplot()
Feature들 사이의 상관관계를 scatterplot으로 나타내는 plot.

**Parameter**: data, hue, palette, size, diag_kind, diag_kws, plot_kws

- `hue`를 기준으로 scatterplot의 색깔이 나뉜다.
- `diag_kind`는 직접 지정해줄 수도 있고 아니면 `hue`에 따라 달라진다.

```python
# 예제
g = sns.pairplot(train[['Survived', 'Pclass', 'Sex', 'Age', 'Parch',
                       'Fare', 'Embarked', 'FamilySize', 'Title']],
                hue='Survived', palette='seismic', size=1.2, diag_kind='kde',
                diag_kws=dict(shade=True), plot_kws=dict(s=10))
g.set(xticklabels=[])
```

```python
import seaborn as sns
%matplotlib inline

iris = sns.load_dataset('iris')
g = sns.pairplot(iris, hue='species', markers=['o', 's', 'D'])
```

```python
g = sns.pairplot(iris, hue='species', markers=['o', 's', 'D'], vars=['sepal_width', 'petal_length'])
```

# seaborn.set()
앞으로 그릴 seaborn 함수에 대한 환경설정.

**Parameter**: context, style, palette

**context**
- 그래프의 테마을 변경해줌. default 설정은 notebook. 이외에 paper, talk, poster 등이 있음.
<br><br>

**style**
- axis 색깔 바뀌고 그런거. 종류에는 darkgrid, whitegrid, dark, white, ticks가 있음.
<br><br>

**palette**
- 그래프 막대기 색상 선정. 종류에는 deep, muted, bright, pastel, dark, colorblind가 있음.
